```python
import statistics

def read_files_and_create_lists(file_paths):
    # Create an empty list for each file
    lists = [[] for _ in range(len(file_paths))]
    results = []

    # Iterate through each file
    for i, file_path in enumerate(file_paths):
        # Open the file
        with open(file_path, 'r') as file:
            # Read each line in the file
            for line in file:
                # Split the line into index and value
                index, value = map(int, line.strip().split(','))

                # Fill the value at the respective index in the list
                lists[i].append(value)

        # Additional processing
        episode_durations = lists[i]
        check = True
        for j in range(11, len(episode_durations)):
            window = episode_durations[j-10:j]
            all_less_than_or_equal_to_10 = all(duration >= -10 for duration in window)
            if all_less_than_or_equal_to_10:
                check = False
                results.append(j)
                break
        if check:
            results.append(500)

    return lists, results
# Example usage
file_paths = []
file_paths2 =  []
file_paths3 = []
seed = 719

path1 = str(seed) + "_PN.txt"
path2 = str(seed) + "_NN.txt"
path3 = str(seed) + "_NN(E).txt"
file_paths.append(path1)
file_paths2.append(path2)
file_paths3.append(path3)

lists, results = read_files_and_create_lists(file_paths)

lists2, results2 = read_files_and_create_lists(file_paths2)

lists3, results3 = read_files_and_create_lists(file_paths3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

def plot_moving_averages(list1, list2, list3):
    moving_avg1 = np.convolve(list1, np.ones(10)/10, mode='valid')
    moving_avg2 = np.convolve(list2, np.ones(10)/10, mode='valid')
    moving_avg3 = np.convolve(list3, np.ones(10)/10, mode='valid')

    # Plot the moving averages
    plt.figure(figsize=(10, 6))
    plt.plot(np.arange(9, len(list1)), moving_avg1, color='blue', label='RYB-DQN-Exp')
    plt.plot(np.arange(9, len(list2)), moving_avg2, color='red', label='DQN')
    plt.plot(np.arange(9, len(list3)), moving_avg3, color='Green', label='RYB period till end')


    # Add labels, title, legend, etc.
    plt.xlabel('Episodes', fontsize=12)
    plt.ylabel('Return', fontsize=12)
    plt.legend(fontsize=12)
    plt.grid(True)

    plt.tick_params(axis='x', labelsize=12)  # Adjust font size for x-axis
    plt.tick_params(axis='y', labelsize=12)  # Adjust font size for y-axis

    plt.show()


ind = 0


plot_moving_averages(lists2[ind], lists[ind],lists3[ind])
```
